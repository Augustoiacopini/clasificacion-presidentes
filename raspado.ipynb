# Scrap de los discursos de Nestor y cristina

```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.cfkargentina.com/category/cfk/discursos/'
```

```python
if __name__ == '__main__':
    response = requests.get(url)
    if response.status_code == 200:
        content = response.text
        with open('pagina_cristina.html', 'w+') as file:
            file.write(content)
        for line in content.split('\n'):
            print(line)
```

```python
with open('pagina_cristina.html', 'r') as file:
    content = file.read()
    soup = BeautifulSoup(content, 'html.parser')
    
    
    
lista = [] 
nombre=[]
link_discursos = soup.find_all(class_ = 'overdefultlink')
for element in link_discursos:
    lista.append(element)
    nombre.append(element.text)
```

```python
link_cristina= []
for l in range(len(lista)):
    link_cristina.append(lista[l]['href'])
```

```python
link_cristina
```

```python
nombre
```

```python
data = pd.DataFrame({'discurso': pd.Series(nombre), 'link':pd.Series(link_cristina)})
data
```

```python
index = data[ data['discurso'] == '\n\n' ].index
# Delete these row indexes from dataFrame
data = data.drop(index)
```

```python
df_cristina_link = data[['discurso', 'link']]
```

```python
df_cristina_link.reset_index
```

```python
df_cristina_link.link.unique()[0]
```

```python
if __name__ == '__main__':
    for l in range(9):
        response = requests.get(df_cristina_link.link.unique()[l])
        if response.status_code == 200:
            content = response.text
            with open(f'cristina{str(l)}.html', 'w+') as file:
                file.write(content)
```
