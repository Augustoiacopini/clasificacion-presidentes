```python
from bs4 import BeautifulSoup

with open('nestor1.html', 'r') as file:
    content = file.read()
    soup = BeautifulSoup(content, 'html.parser')
    for line in content.split('\n'):
            print(line)
```

```python
texto = soup.find_all('p')
parrafo = []
for element in texto:
    parrafo.append(element.text)
```

```python
parrafo.remove('\xa0')
parrafo.remove(parrafo[0])
```
